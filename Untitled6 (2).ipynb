```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/content/drive/MyDrive/students.csv")
df
```

```python
print(df.isnull().sum())
df = df.dropna()
print(df)
```

```python
marks = df["Marks"]
mean_marks = np.mean(marks)
print("Mean =", mean_marks)
median_marks = np.median(marks)
print("Median =", median_marks)
std_marks = np.std(marks)
print("Standard Deviation =", std_marks)
```

```python
plt.hist(df["Marks"], bins=5)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram Marks")
plt.show()
```

```python
plt.scatter(df["Study Hours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
X = df[["Study Hours"]]
y = df["Marks"]
model = LinearRegression()
model.fit(X, y)
```

```python
predicted = model.predict(X)
print("Predicted Values:")
print(predicted)
```

```python
plt.scatter(df["Study Hours"], df["Marks"], label="Actual Data")
plt.plot(df["Study Hours"], predicted, color="red", label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
